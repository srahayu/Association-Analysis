## association analis

association analysis digunakan untuk mengetahui produk yang berpotensi untuk dikombinasikan dengan produk lain, sehingga bisa diberikan strategi pemasaran berupa bundling produk

```python
pip install mlxtend 
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
bakery_item=bakery.reset_index(inplace=True)
bakery_item=bakery.drop(['date','place','day of week','total','dates','month','day','hour','time','index'], axis=1, index=None)
bakery_item=bakery_item >=1
bakery_item.head()
```

```python
bakery_apriori=apriori(bakery_item,min_support =0.01, use_colnames=True)
bakery_apriori.sort_values(by='support',ascending=False).head()
```

```python
bakery_apriori['length']=bakery_apriori['itemsets'].apply(lambda x : len(x))
```

```python
bakery_apriori2=bakery_apriori[(bakery_apriori['length']==2)& bakery_apriori['support']>=0.05]
bakery_apriori2.sort_values(by='support',ascending=False).head()
```

```python
rules=association_rules(bakery_apriori,metric='lift',min_threshold=1)
rules.sort_values(by='lift',ascending=False, inplace=True)
rules.head()
```

#### conclusion

```python
rules2=association_rules(bakery_apriori,metric='lift',min_threshold=1)
rules2.sort_values(by='confidence',ascending=False, inplace=True)
rules2.head()
```
